# IPL Data Analysis - Final Project


```python
import pandas as pd

# Load both CSVs
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")

# Preview
print("Matches Data:")
display(matches.head())

print("Deliveries Data:")
display(deliveries.head())
```

```python
# Group by batsman and sum runs
top_scorers = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)

# Plotting
import matplotlib.pyplot as plt

top_scorers.plot(kind='bar', figsize=(10,6), color='orange')
plt.title("Top 10 Run Scorers in IPL History")
plt.xlabel("Batsman")
plt.ylabel("Total Runs")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
